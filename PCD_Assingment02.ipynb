```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Upload an image file
uploaded = files.upload()

# Get the filename and read the image in color
filename = next(iter(uploaded))
original_image = cv2.imread(filename)

# Convert from BGR to RGB for displaying with matplotlib
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)

print(f"\nDisplaying '{filename}'")

# Display the original image
plt.figure(figsize=(5, 5))
plt.imshow(original_image_rgb)
plt.title("Original Image")
plt.axis('off')
plt.show()
```

```python
# Apply a Gaussian blur to the image
# The (7, 7) is the kernel size, which determines the amount of blur
blurred_image = cv2.GaussianBlur(original_image, (7, 7), 0)

# Convert images to RGB for displaying
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
blurred_image_rgb = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB)

# Plot the original and blurred images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(original_image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Gaussian Blur")
plt.imshow(blurred_image_rgb)
plt.axis('off')
plt.show()
```

```python
# Set the brightness increase value
brightness_value = 50

# Create a matrix of ones with the same shape as the image, then multiply by the brightness value
brightness_matrix = np.ones(original_image.shape, dtype="uint8") * brightness_value

# Add the brightness matrix to the original image
bright_image = cv2.add(original_image, brightness_matrix)

# Convert images to RGB for displaying
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
bright_image_rgb = cv2.cvtColor(bright_image, cv2.COLOR_BGR2RGB)

# Plot the original and brightened images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(original_image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Brightened Image")
plt.imshow(bright_image_rgb)
plt.axis('off')
plt.show()
```

```python
# Set the darkness value
darkness_value = 50

# Create a matrix of ones and multiply by the darkness value
darkness_matrix = np.ones(original_image.shape, dtype="uint8") * darkness_value

# Subtract the darkness matrix from the original image
dark_image = cv2.subtract(original_image, darkness_matrix)

# Convert images to RGB for displaying
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
dark_image_rgb = cv2.cvtColor(dark_image, cv2.COLOR_BGR2RGB)

# Plot the original and darkened images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(original_image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Darkened Image")
plt.imshow(dark_image_rgb)
plt.axis('off')
plt.show()
```

```python
# Alpha controls contrast (0.0-1.0 for lower, >1.0 for higher)
alpha = 0.5
# Beta controls brightness (0-100)
beta = 50

# Apply the contrast adjustment
low_contrast_image = cv2.addWeighted(original_image, alpha, np.zeros(original_image.shape, original_image.dtype), 0, beta)

# Convert images to RGB for displaying
original_image_rgb = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)
low_contrast_rgb = cv2.cvtColor(low_contrast_image, cv2.COLOR_BGR2RGB)

# Plot the original and low-contrast images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(original_image_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Lower Contrast")
plt.imshow(low_contrast_rgb)
plt.axis('off')
plt.show()
```
